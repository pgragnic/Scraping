```python
import requests
import pandas as pd
import time
```

```python
df = pd.DataFrame()
page = "https://www.data.gouv.fr/api/1/datasets/?page=1&page_size=200"
i = 0
csv = 0

while page is not None:
    i += 1
    print("Step " + str(i) + ", page: " + page)
    start_time = time.time()
    r = requests.get(page)
    data = r.json()
    # dfOne = pd.json_normalize(data)
    dfOne = pd.DataFrame.from_dict(data['data'])
    df = df.append(dfOne)
    page = data["next_page"]
    if ((i%5) == 0):
        csv += 1
        print(f"Write file {str(csv)}")
        df.to_csv(f"test{csv}.csv", sep=";")
        df = pd.DataFrame()
    print("It took %s seconds ---" % (time.time() - start_time))

df.to_csv(f"test-global.csv", sep=";")
```
